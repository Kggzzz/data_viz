```
import pandas as pd 
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

#Parallel Coordinates

ถ้ามีข้อมูลน้อย ใช้ง่ายกว่า

```
pd.plotting.parallel_coordinates(df, 'Name')
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=('#556270', '#4ECDC4', '#C7F464'))
#เรียกแบบ HTML color เป็นเลขฐาน16
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=('r', 'g', 'b'))
#เรียกแบบตัวย่อของสี
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=((1,0,0),(0,1,0),(0,0,1)))
#เรียกใช้แบบ RGB value (r,g,b)
#ซึ่ง r g b  จะต้องมีค่าตั้งแต่ 0 - 1
```

## ปรับ Parameter : Colormap

Colormap เป็นสีที่คนคิดมาแล้วว่า มีสีไหนที่เข้ากันแล้วบ้าง เป็นชุดสีออกมาโดยเราไม่ต้องกำหนดสีเอง 

```
pd.plotting.parallel_coordinates(df, 'Name', colormap = 'cool')
```

```
pd.plotting.parallel_coordinates(df, 'Name', colormap = 'Set3')
```

#เริ่ม Matplotlib

```
from matplotlib import pyplot as plt  
```

##แปลง Data จาก pandas เป็น  numpy

```
np_data = df.iloc[:,:-1].to_numpy()
```

#Visualize array data กับ pixle

ถ้ามีข้อมูลเยอะใช้ pixel ดูง่ายกว่า 

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[15,15] #ยิ่งใส่เลขมาก ภาพก็จะยิ่งใหญ่มาก 
plt.imshow(np_data[:10,:],interpolation='nearest')
```

###การใช้ subplot กับ Imshow 

```
plt.subplot(1,3,1)
plt.imshow(np_data[:50,:],interpolation='nearest')
plt.subplot(1,3,2)
plt.imshow(np_data[50:100,:],interpolation='nearest')
plt.subplot(1,3,3)
plt.imshow(np_data[100:150,:],interpolation='nearest')
```

#alpha ความใสของสี

```
plt.imshow(np_data[0:10,:],interpolation='nearest',alpha = 0.5)
```
